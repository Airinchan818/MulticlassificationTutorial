```python
from sklearn.datasets import make_multilabel_classification,make_regression
```

```python
from littlelearn import DeepLearning as dl
import littlelearn as ll
```

```python
x_train,y_train = make_multilabel_classification(n_classes=4,n_labels=8)
```

```python
y_train = y_train.reshape((y_train.shape[0],2,2))
```

```python
class NeuralNetwork :
  def __init__(self) :
    self.linear1 = dl.layers.Dense(32,activation='relu')
    self.linear2 = dl.layers.Dense(64,activation='relu')
    self.linear3 = dl.layers.Dense(128,activation='relu')
    self.linear4 = dl.layers.Dense(256,activation='relu')
    self.out = dl.layers.Dense(8)
    self.node = [self.linear1,self.linear2,self.linear3,self.linear4,self.out]

  def get_weight(self) :
    weight = list()
    for node in self.node :
      wg = node.get_weight()
      if wg is not None :
        for w in wg :
          weight.append(w)
    return weight

  def __call__ (self,x) :
    x = self.linear1(x)
    x = self.linear2(x)
    x = self.linear3(x)
    x = self.linear4(x)
    x =  self.out(x)
    return x.reshape((x.shape[0],4,2))
```

```python
model = NeuralNetwork()
model(x_train[:1].copy())
loss_fn = dl.loss.SparseCategoricallCrossentropy(from_logits=True)
optimizersfn = dl.optimizers.Adam()
optimizersfn.apply_weight(model.get_weight())
clipper = ll.GradientTools.ClipByNorm(model.get_weight(),max_norm=1.0)
hist = list()
```

```python
epochs = 100
for epoch in range(epochs) :
  y_pred = model(x_train)
  loss = loss_fn(y_train,y_pred)
  loss.backwardpass()
  clipper.execute()
  optimizersfn.forward_in_weight()
  loss.kill_grad()
  hist.append(loss.get_tensor())
  print(f"epoch : {epoch + 1} / {epochs} || loss : {loss.get_tensor():.5f}")
```

```python
clipper.plot_gradient_magnitude()
```

```python
import matplotlib.pyplot as plt
plt.title("loss")
plt.xlabel('epoch per step')
plt.ylabel("values")
plt.plot(hist,label='loss',color='red')
plt.legend()
plt.grid(True)
plt.show()
```
